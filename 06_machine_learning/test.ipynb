```python
import numpy as np
import matplotlib.pyplot as plt

#  
positions = np.arange(0, 50)   # 0~49   

#     3 sin 
freqs = [1, 0.1, 0.01]  #   (   ,    )

plt.figure(figsize=(10, 4))
for f in freqs:
    plt.plot(positions, np.sin(positions * f), label=f"sin(pos * {f})")

plt.title("Positional Encoding: Different Frequencies (sin)")
plt.xlabel("Position (token index)")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

positions = np.arange(0, 50)

freqs = [1, 0.1, 0.01]

plt.figure(figsize=(10, 6))

for f in freqs:
    plt.plot(positions, np.sin(positions * f), label=f"sin(pos * {f})")

for f in freqs:
    plt.plot(positions, np.cos(positions * f), linestyle='--', label=f"cos(pos * {f})")

plt.title("Sine & Cosine Positional Encoding with Different Frequencies")
plt.xlabel("Position Index (token index)")
plt.ylabel("Value")
plt.legend()
plt.show()
```
