```python
# CSV íŒŒì¼ ì½ì–´ì„œ DBì— ì €ì¥í•˜ê¸°
import pandas as pd
import pymysql

# MySQL ì—°ê²°
conn = pymysql.connect(
    host='127.0.0.1',
    user='root',
    password='1111',
    db='1st',
    port=3306,
    charset='utf8'
)

try:
    cursor = conn.cursor()
    
    # í…Œì´ë¸” ìƒì„±
    create_sql = """
    CREATE TABLE emergency_car (
        idx INT AUTO_INCREMENT PRIMARY KEY,
        year YEAR NOT NULL,
        car_count INT,
        emp_count INT,
        car_local VARCHAR(50) NOT NULL
    );
    """
    # ê¸°ì¡´ í…Œì´ë¸” ì‚­ì œ í›„ ìƒˆë¡œ ìƒì„±
    cursor.execute("DROP TABLE IF EXISTS emergency_car")
    cursor.execute(create_sql)
    
    # CSV íŒŒì¼ ì½ê¸°
    df = pd.read_csv('../csv/emergency_car.csv')  # CSV íŒŒì¼ ê²½ë¡œë¥¼ ì‹¤ì œ íŒŒì¼ ìœ„ì¹˜ë¡œ ìˆ˜ì •í•´ì£¼ì„¸ìš”
    
    # DataFrameì„ ì´ìš©í•´ì„œ ë°ì´í„° ì‚½ì…
    for index, row in df.iterrows():
        sql = """
        INSERT INTO emergency_car (year, car_count, emp_count, car_local)
        VALUES (%s, %s, %s, %s)
        """
        values = (row['year'], row['car_count'], row['emp_count'], row['car_local'])
        cursor.execute(sql, values)
    
    conn.commit()
    print("CSV ë°ì´í„°ê°€ ì„±ê³µì ìœ¼ë¡œ ë°ì´í„°ë² ì´ìŠ¤ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤!")

except Exception as e:
    print("ì˜¤ë¥˜ ë°œìƒ:", e)
    conn.rollback()

finally:
    cursor.close()
    conn.close()
```
