```python
import pymysql
import pandas as pd

# MySQL 연결
conn = pymysql.connect(
    host='127.0.0.1',
    user='root',
    password='1111',
    db='1st',
    port=3306,
    charset='utf8'
)

try:
    # 특정 지역 데이터 조회
    region = input("조회할 지역을 입력하세요 (예: 서울, 부산, 대구 등): ")
    sql_region = f"""
    SELECT year, car_local, car_count, emp_count
    FROM emergency_car
    WHERE car_local = '{region}'
    ORDER BY year;
    """
    df_region = pd.read_sql(sql_region, conn)
    
    if len(df_region) > 0:
        print(f"\n📍 {region} 지역 응급차량 데이터")
        print(df_region)
        print(f"\n{region} 지역 데이터 수:", len(df_region))
    else:
        print(f"\n⚠️ {region} 지역의 데이터가 없습니다.")

finally:
    conn.close()  # 연결 종료
```

```python
# CSV 파일 읽어서 DB에 저장하기
import pandas as pd
import pymysql

# MySQL 연결
conn = pymysql.connect(
    host='127.0.0.1',
    user='root',
    password='1111',
    db='1st',
    port=3306,
    charset='utf8'
)

try:
    cursor = conn.cursor()
    
    # 테이블 생성
    create_sql = """
    CREATE TABLE emergency_car (
        idx INT AUTO_INCREMENT PRIMARY KEY,
        year YEAR NOT NULL,
        car_count INT,
        emp_count INT,
        car_local VARCHAR(50) NOT NULL
    );
    """
    # 기존 테이블 삭제 후 새로 생성
    cursor.execute("DROP TABLE IF EXISTS emergency_car")
    cursor.execute(create_sql)
    
    # CSV 파일 읽기
    df = pd.read_csv('../csv/emergency_car.csv')  # CSV 파일 경로를 실제 파일 위치로 수정해주세요
    
    # DataFrame을 이용해서 데이터 삽입
    for index, row in df.iterrows():
        sql = """
        INSERT INTO emergency_car (year, car_count, emp_count, car_local)
        VALUES (%s, %s, %s, %s)
        """
        values = (row['year'], row['car_count'], row['emp_count'], row['car_local'])
        cursor.execute(sql, values)
    
    conn.commit()
    print("CSV 데이터가 성공적으로 데이터베이스에 저장되었습니다!")

except Exception as e:
    print("오류 발생:", e)
    conn.rollback()

finally:
    cursor.close()
    conn.close()
```
